```python
opcion_menu = ""

# Ciclo while para mantener el menú activo hasta que se elija salir [2]
while opcion_menu != "4":
    print("\n--- ⚒️ IA HERRERA: ANÁLISIS DE FORJA ---")
    print("1. Tipo de metal: COBRE (Blando)")
    print("2. Tipo de metal: HIERRO (Medio)")
    print("3. Tipo de metal: ACERO (Duro)")
    print("4. Apagar fragua y salir")

    # Captura de la opción del usuario
    opcion_menu = input("Elija un metal para analizar su forjado: ")

    # Estructura condicional para validar opciones de metal [3]
    if opcion_menu == "1" or opcion_menu == "2" or opcion_menu == "3":
        # Entrada de datos para el cálculo lógico
        calor = float(input("Nivel de calor en la fragua (1-100): "))
        martillazos = float(input("Cantidad de golpes de fuerza aplicados: "))

        # Validación para evitar división por cero [4]
        if martillazos > 0:
            # Cálculo de la relación de forjado (simulando lógica de IA)
            calidad = calor / martillazos

            # Toma de decisiones basada en el resultado del cálculo [5]
            if calidad > 3:
                print(">> RESULTADO: CALIDAD LEGENDARIA. ¡El metal ha fluido perfectamente!")
            elif calidad >= 1:
                print(">> RESULTADO: CALIDAD ESTÁNDAR. El arma es funcional pero básica.")
            else:
                print(">> RESULTADO: CALIDAD DEFECTUOSA. ¡El metal se quebró por falta de calor!")
        else:
            print("Error: Los martillazos deben ser mayores a cero para forjar.")

    # Opción para terminar el programa [3]
    elif opcion_menu == "4":
        print("Desconectando Sistema Experto del Herrero...")

    # Manejo de opciones no válidas
    else:
        print("Opción no reconocida por la IA.")

# Mensaje final al romper el ciclo
print("¡Taller cerrado!")
```
